**Выбрать тест и проверить, есть  ли различия между выборками:**

**1 )  Даны две  независимые выборки. Не соблюдается условие нормальности**

x1  380,420, 290

y1 140,360,200,900

Сделайте вывод по результатам, полученным с помощью функции 

```python
import numpy as np
import scipy.stats as stats
```

```python
# Две группы. Независимые выборки
# Применим Критерий Манна -Уитни
x1 = np.array([380,420,290])
y1 = np.array([140,360,200,900])
stats.mannwhitneyu(x1, y1)
```

Вывод: pvalue > 0.05 Статистически значимых различий не обнаружено на уровне значимости alpha =0,05. Верна нулевая гипотеза.

**2 ) Исследовалось влияние препарата на уровень давления пациентов.** 

Сначала измерялось давление до приема препарата, потом через 10 минут и через 30 минут. Есть ли статистически значимые различия?

1е измерение до приема препарата: 150, 160, 165, 145, 155

2е измерение через 10 минут: 140, 155, 150,  130, 135

3е измерение через 30 минут: 130, 130, 120, 130, 125

```python
# Множественные сравнения. Анализ повторных измерений
# Применим Критерий Фридмана
dimension_1 = np.array([150, 160, 165, 145, 155])
dimension_2 = np.array([140, 155, 150, 130, 135])
dimension_3 = np.array([130, 130, 120, 130, 125])
stats.friedmanchisquare(dimension_1, dimension_2, dimension_3)
```

Вывод: pvalue < 0.05. Присутствуют статистически значимые различия. Принимается альтернативная гипотеза. Препарат влияет на уровень давления.

**3 ) Сравните 1 и 2 е измерения, предполагая, что 3го измерения через 30 минут не было.** (из второй задачи)

```python
# Српавнение двух групп. Зависимые выборки
# Применим критерий Уилкоксона
dimension_1 = np.array([150, 160, 165, 145, 155])
dimension_2 = np.array([140, 155, 150, 130, 135])
stats.wilcoxon(dimension_1,dimension_2)
```

Вывод: pvalue > 0.05. Cтатистически значимых различий нет. Препарат НЕ влияет на уровень давления.

**4) Даны 3 группы  учеников плавания.**

В 1 группе время на дистанцию 50 м составляют:

56, 60, 62, 55, 71, 67, 59, 58, 64, 67

Вторая группа : 57, 58, 69, 48, 72, 70, 68, 71, 50, 53

Третья группа: 57, 67, 49, 48, 47, 55, 66, 51, 54

```python
# Множественные сравнения. Независимые выборки
# Применим Критерий Краскала-Уоллиса
gr_1 = np.array([56, 60, 62, 55, 71, 67, 59, 58, 64, 67])
gr_2 = np.array([57, 58, 69, 48, 72, 70, 68, 71, 50, 53])
gr_3 = np.array([57, 67, 49, 48, 47, 55, 66, 51, 54])
stats.kruskal(gr_1, gr_2, gr_3)
```

**5) Заявляется, что партия изготавливается со средним арифметическим 2,5 см.** Проверить данную гипотезу, если известно, что размеры изделий подчинены нормальному закону распределения. Объем выборки 10, уровень статистической значимости 5%

2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34

H_0: М = 2,5

Н_1: М ≠ 2,5

alpha =0,05

Применим Z критерий


```python

#найдем среднее арифметическое
X = np.array([2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34])
mean = X.mean()
mean
```

```python
# Найдем стандартное отклонение
std = X.std(ddof=1)
std
```

```python
# Вычислим t статистику
t_fact = (mean - 2.5) / std * np.sqrt(len(X))
t_fact
```

По таблице для критерия Стьюдента находим t = 1.833 (alpha = 0.05 , 9 степеней свободы).

```python
alpha = 0.05
n = (len(X))

t_table = stats.t.ppf(1 - alpha, n-1)
t_table
```

Вывод: Так как t_table > t_fact, то нулевая гипотеза верна. Размер изделий со средним арифметическим 2,5 см
